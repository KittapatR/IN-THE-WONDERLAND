```python
import glob
import geopandas as gpd

paths = glob.glob("Final display/*.shp")
```

```python
import pandas as pd
ref_province = pd.read_excel("จังหวัด.xlsx", sheet_name = "ref_label_province")
ref_province
```

```python
for path in paths:
    tmp = gpd.read_file(path, encoding = "utf-8")
    tmp["group"]= (abbrev := ref_province[ref_province["จังหวัด"] == tmp["P_name"][0]]["อักษรย่อโรมัน[2]"].to_list()[0])
    tmp["CONS_id"] = tmp.apply(lambda x: x["group"] + "_" + str(x["CONS_no"]) + "_area", axis = 1)
    tmp.to_file("CONS_{}.shp".format(abbrev))
```
